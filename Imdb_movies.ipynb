```python
# Convert Gross column to integers using apply and lambda
df['Gross'] = df['Gross'].apply(lambda x: int(x.replace(',', '')) if isinstance(x, str) else x)

# Convert Runtime to numeric format (removing ' min')
df['Runtime'] = pd.to_numeric(df['Runtime'].str.replace(' min', ''), errors='coerce')

# Define bins for runtime ranges
runtime_bins = [0, 90, 120, 150, 180, float('inf')]
runtime_labels = ['<90', '90-120', '120-150', '150-180', '180+']

# Categorize movies into runtime ranges
df['Runtime Range'] = pd.cut(df['Runtime'], bins=runtime_bins, labels=runtime_labels)

# Calculate average gross earnings for each runtime range
avg_gross_by_runtime = df.groupby('Runtime Range')['Gross'].mean()

# Create a bar chart to compare average gross earnings for different runtime ranges
plt.figure(figsize=(10, 6))
avg_gross_by_runtime.plot(kind='bar', color='skyblue')
plt.xlabel('Runtime Range')
plt.ylabel('Average Gross Earnings')
plt.title('Average Gross Earnings by Runtime Range')
plt.xticks(rotation=45)
plt.tight_layout()

# Display the plot
plt.show()
```

```python
grouped_data = df.groupby('Genre').agg({'IMDB_Rating': 'mean', 'Runtime': 'mean'}).reset_index()
plt.figure(figsize=(10, 6))

# Size of bubbles based on the number of movies in each genre
bubble_sizes = df['Genre'].value_counts() * 5

plt.scatter(
    x=grouped_data['IMDB_Rating'],
    y=grouped_data['Runtime'],
    s=bubble_sizes,
    alpha=0.5
)

# Adding labels and title
plt.xlabel('IMDB Rating')
plt.ylabel('Runtime')
plt.title('IMDB Rating vs. Runtime of Movies by Genre')

# Adding text labels for genres
for i, row in grouped_data.iterrows():
    plt.text(row['IMDB_Rating'], row['Runtime'], row['Genre'], fontsize=8, ha='center', va='center')

plt.grid(True)
plt.show()

```

```python


# Combine the star columns into a single 'Stars' column
df['Stars'] = df[['Star1', 'Star2', 'Star3', 'Star4']].apply(lambda row: ', '.join(row.dropna()), axis=1)

# Group by actor combinations and calculate the average gross for each combination
actor_gross = df.groupby('Stars')['Gross'].mean().reset_index()

# Sort by average gross in descending order to get top combinations
top_actor_combinations = actor_gross.sort_values(by='Gross', ascending=False).head(10)

# Create a bar chart for the top actor combinations
plt.figure(figsize=(10, 6))
plt.barh(top_actor_combinations['Stars'], top_actor_combinations['Gross'], color='skyblue')
plt.xlabel('Average Gross')
plt.ylabel('Actor Combinations')
plt.title('Top Ten Actor Combinations with Good Gross')
plt.gca().invert_yaxis()  # Invert y-axis to have the highest gross at the top
plt.show()
```

```python
df['Released_Year'] = pd.to_datetime(df['Released_Year'], format='%Y')

# Calculate average IMDb ratings and total gross earnings per year
average_ratings_per_year = df.groupby(df['Released_Year'].dt.year)['IMDB_Rating'].mean()
total_earnings_per_year = df.groupby(df['Released_Year'].dt.year)['Gross'].sum()

# Create a line chart for average IMDb ratings over the years
plt.figure(figsize=(10, 6))
plt.plot(average_ratings_per_year.index, average_ratings_per_year.values, marker='o')
plt.xlabel('Year')
plt.ylabel('Average IMDb Rating')
plt.title('Average IMDb Ratings Over the Years')
plt.grid(True)
plt.show()

# Create a line chart for total gross earnings over the years
plt.figure(figsize=(10, 6))
plt.plot(total_earnings_per_year.index, total_earnings_per_year.values, marker='o', color='orange')
plt.xlabel('Year')
plt.ylabel('Total Gross Earnings')
plt.title('Total Gross Earnings Over the Years')
plt.grid(True)
plt.show()






```

```python
# Load your dataset
# Assuming your dataset is loaded into a DataFrame named 'df'

# Clean the data by splitting the Genre column into separate genres


# Create a new DataFrame to hold the counts of genres for each year
genre_year_counts = df.explode('Genre').groupby(['Released_Year', 'Genre']).size().unstack(fill_value=0)

# Create a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(genre_year_counts,cmap="YlGnBu", annot=True, fmt='0.5%')
plt.title('Genre Distribution Over Released Years')
plt.xlabel('Genre')
plt.ylabel('Released Year')
plt.show()
```

```python
top_10_genres = df['Genre'].explode().value_counts().head(10).index

# Create a new DataFrame to hold the total earnings of top 10 genres for each year
top_10_genre_year_earnings = df[df['Genre'].explode().isin(top_10_genres)].explode('Genre').groupby(['Released_Year', 'Genre'])['Gross'].sum().unstack(fill_value=0)

# Create a line chart
plt.figure(figsize=(12, 8))
for genre in top_10_genres:
    plt.plot(top_10_genre_year_earnings.index, top_10_genre_year_earnings[genre], label=genre, color=sns.color_palette()[top_10_genres.get_loc(genre)])
plt.title('Genre Earnings Over Released Years (Top 10)')
plt.xlabel('Released Year')
plt.ylabel('Total Earnings')
plt.legend()
plt.show()
```
