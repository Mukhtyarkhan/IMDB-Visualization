```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r'D:\Data Sicence bootcamp\activity datastorytelling\imdb_top_1000.csv')
```

```python
df.head(5)
```

```python
# Clean the data by removing non-numeric values from 'Released_Year'
df['Released_Year'] = df['Released_Year'].str.extract(r'(\d{4})')

# Convert 'Released_Year' to numeric values
df['Released_Year'] = pd.to_numeric(df['Released_Year'], errors='coerce')

# Drop rows with missing 'Released_Year' values
df.dropna(subset=['Released_Year'], inplace=True)

# Group data by year and count the number of movies released each year
movies_per_year = df['Released_Year'].value_counts().sort_index()

# Create a line graph for the number of movies released each year
plt.figure(figsize=(10, 6))
plt.plot(movies_per_year.index, movies_per_year.values, marker='o', linestyle='-', color='b')
plt.xlabel('Year')
plt.ylabel('Number of Movies Released')
plt.title('Number of Movies Released Each Year')
plt.grid(True)
plt.tight_layout()

# Display the plot
plt.show()
```

```python
genra_counts=df['Genre'].str.split(',').explode().value_counts()
```
