```python
# Data
data = {'A': 30, 'B': 20, 'C': 50}

# Create a pie chart
plt.pie(data.values(), labels=data.keys(),autopct='%1.1f%%')

# Add a title
plt.title('Pie Chart Example')

# Display the plot
plt.show()
```

```python
# Data and DataFrame
data = {'Category': ['A', 'B', 'C'],
        'Value': [30, 20, 50]}
df = pd.DataFrame(data)

# Create a pie chart using Seaborn
sns.catplot(x='Value', y='Category', data=df, kind='pie')

# Add a title
plt.title('Pie Chart Example')

# Display the plot
plt.show()
```
